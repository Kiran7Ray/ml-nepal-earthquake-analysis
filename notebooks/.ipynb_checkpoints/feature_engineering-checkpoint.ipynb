```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("../data/cleaned_data.csv")
data.head()
```

convert time column

```python
print(data['DateTime'])
```

```python
data['DateTime']=pd.to_datetime(data['DateTime'])
data['DateTime']
```

extract year

```python
data['year']=data['DateTime'].dt.year
data['year']
```

extract month

```python
data['month']=data['DateTime'].dt.month
data['month']
```

extract day

```python
data['day'] = data['DateTime'].dt.day
data['day']
```

extract hour,weekend

```python
data['hour']=data['DateTime'].dt.hour
data['weekday'] = data['DateTime'].dt.dayofweek
```

```python
data.head()
```

keeping useful numerical feature

```python
features = [
    'latitude',
    'magnitude',
    'longitude',
    'depth',
    'nst',
    'gap',
    'dmin',
    'rms',
    'horizontalError',
    'depthError',
    'magError',
    'magNst',
    'year',
    'month',
    'day',
    'hour',
    'weekday'
]
```

handle categorical columns

```python
data.columns
```

```python
for cols in ['magnitudeType','type','place']:
    
    print(data[cols].value_counts())
```

one hot encoding

```python
data.columns
```

```python
cat_cols = [
    'magnitudeType',
    'type',
    'place'
]

data = pd.get_dummies(data, columns=cat_cols, drop_first=True)
```

```python
data.head()
```

nepal earthquack often cluster geographically ,create resional feature

```python
def region(lat):
    
    if lat < 27.5:
        return "South"
    
    elif lat < 29:
        return "Central"
    
    else:
        return "North"

data['region'] = data['latitude'].apply(region)
```

```python
data = pd.get_dummies(
    data,
    columns=['region'],
    drop_first=True
)
```

feature selection

```python
corr = data.corr(numeric_only=True)

corr['magnitude'].sort_values(
    ascending=False
)
```
