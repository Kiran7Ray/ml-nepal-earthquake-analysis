```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../data/nepal_earthquakes_1990_2026.csv")
```

Duplicate row

```python
duplicate=df.duplicated().sum()
print("duplicate rows:",duplicate)
```

```python
print(df.isnull().sum())
```

Missing value treatment

```python
num_cols=df.select_dtypes(include=np.number).columns
for col in num_cols:
    df[col]=df[col].fillna(df[col].median())
```

```python
print(df.isnull().sum())
```

rename column name (short form in long form)

```python
df=df.rename(columns={'time':'DateTime','mag':'magnitude','magType':'magnitudeType'})
```

```python
df=df.drop(columns=['id','updated','net','status','locationSource','magSource'])
```

```python
print(df['place'].unique())
```

```python
problem_places = df[df['place'].str.contains(r'\?')]

print(problem_places['place'].unique())
```

```python
place_corrections = {
    "B?glung": "Baglung",
    "Dh?rchula": "Dharchula",
    "D?rchul?": "Darchula",
    "Kod?ri??": "Kodari",
    "Kh??db?ri??": "Khandbari",
    "Il?m": "Ilam",
    "Dhankut?": "Dhankuta",
    "P?tan": "Patan",
    "Panauti??": "Panauti",
    "R?mnagar": "Ramnagar",
    "B?geshwar": "Bageshwar",
    "Naya B?z?r": "Naya Bazar",
    "Dadeldhur?": "Dadeldhura",
    "Josh?math": "Joshimath",
    "Dhar?n": "Dharan",
    "?ik?pur": "Tikapur",
    "D?rjiling": "Darjiling",
    "Kh??db?ri": "Khandbari",
    "R?jbir?j": "Rajbiraj",
    "Lohagh?t": "Lohaghat",
    "Tuls?pur": "Tulsipur",
    "Pithor?garh": "Pithoragarh",
    "Lah?n":"Lahan",
"Odl?b?ri":"Odlabari"
}
```

```python
for wrong, correct in place_corrections.items():
    df['place'] = df['place'].str.replace(
        wrong,
        correct,
        regex=False
    )
```
